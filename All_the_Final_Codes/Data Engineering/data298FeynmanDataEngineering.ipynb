```python
import requests
from bs4 import BeautifulSoup

# URL of the dataset
url = 'https://www.aip.org/history-programs/niels-bohr-library/oral-histories/5020-1'

response = requests.get(url)

if response.status_code == 200:
    soup = BeautifulSoup(response.text, 'html.parser')
    elements = soup.find_all('div', class_='field-item even')

    # write the extracted content
    with open('/content/drive/MyDrive/DATA298B/data/interview_transcript.txt', 'w', encoding='utf-8') as file:
        for element in elements:
            file.write(element.get_text(separator='\n', strip=True))
            file.write('\n\n')
else:
    print("Failed to retrieve the webpage. Status code:", response.status_code)
```

```python
with open('/content/drive/MyDrive/DATA298B/data/interview_transcript.txt', 'r') as file:
    lines = file.readlines()

# To store Weiner's questions and Feynman's answers
questions = []
answers = []

# Keep track of who is speaking
current_speaker = None
current_text = ""

for line in lines:
    line = line.strip()

    if line.startswith('Weiner:'):
        if current_speaker == 'Feynman':
            answers.append(current_text.strip())

        current_speaker = 'Weiner'
        current_text = line.replace('Weiner:', '').strip()

    elif line.startswith('Feynman:'):
        if current_speaker == 'Weiner':
            questions.append(current_text.strip())

        current_speaker = 'Feynman'
        current_text = line.replace('Feynman:', '').strip()

    else:
        current_text += ' ' + line.strip()

if current_speaker == 'Weiner':
    questions.append(current_text.strip())
elif current_speaker == 'Feynman':
    answers.append(current_text.strip())

print("Questions by Weiner:")
print(questions)
print("\nAnswers by Feynman:")
print(answers)
```

```python
# check the data
num_questions = len(questions)
num_answers = len(answers)

print("Number of questions by Weiner:", num_questions)
print("Number of answers by Feynman:", num_answers)
```

```python
for question, answer in zip(questions, answers):
    print(f"Q: {question}")
    print(f"A: {answer}")
    print("---")
```

```python
# Open a text file to write the Q&A pairs
with open('/content/drive/MyDrive/DATA298B/data/interview_transcript_Q&A.txt', 'w', encoding='utf-8') as output_file:
  for question, answer in zip(questions, answers):
    output_file.write(f"Q: {question}\n")
    output_file.write(f"A: {answer}\n")
    output_file.write("\n")
```

```python
def split_dialogue_by_empty_lines(conversation):
    dialogue_units = conversation.strip().split('\n\n')
    return dialogue_units
```

```python
file_path = 'interview_transcript_QA.txt'

with open(file_path, 'r', encoding='utf-8') as file:
    texts = file.read()

split_dialogues = split_dialogue_by_empty_lines(texts)

for idx, dialogue in enumerate(split_dialogues, 1):
    print(f"Dialogue {idx}:")
    print(dialogue)
    print("-----")
```

```python
documents = [Document(page_content=dialogue) for dialogue in split_dialogues]
print(documents)
```
